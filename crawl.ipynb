```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
BASEURL = 'https://thuvienphapluat.vn'
```

```python
WebMapURL = '/hoi-dap-phap-luat/so-do-website.html'
WebMap = requests.get(BASEURL + WebMapURL)
WebMapSoup = BeautifulSoup(WebMap.text, 'html.parser')
```

```python
content = WebMapSoup.find('div', class_='dropdown-content')
links = content.find_all('a')
links = [link.get('href') for link in links]
print(links)
```

```python
MainContent = WebMapSoup.find('ul', id = 'myUL')
MainLinks = MainContent.find_all('li')
MainLinks = [link.find('span').find('a').get('href') for link in MainLinks]
print(MainLinks)
```

```python
MainLinks = MainLinks + links
print(MainLinks)
```

```python
def Get_URL_page():
    for i in MainLinks:
        FULL_URL = BASEURL + i
        request_to_list_page = requests.get(FULL_URL)
        soup = BeautifulSoup(request_to_list_page.text, 'html.parser')
        URL_to_last_page = soup.find('ul', class_='pagination').find_all('li')[-1].find('a').get('href')
        MAX_PAGE = int(URL_to_last_page.split('=')[-1])
        for page in range(1, MAX_PAGE + 1):
            yield FULL_URL + '?page=' + str(page)

def Get_URL_article():
    for page in Get_URL_page():
        request_to_list_page = requests.get(page)
        soup = BeautifulSoup(request_to_list_page.text, 'html.parser')
        articles = soup.find_all('div', class_='news-card')
        for article in articles:
            yield BASEURL + article.find('a').get('href')

def Get_article():
    for url in Get_URL_article():
        request_to_article = requests.get(url)
        soup = BeautifulSoup(request_to_article.text, 'html.parser')
        main_content = soup.find('div', id='news-detail')
        data = []
        title = main_content.find('header').find('h1').text
        content = main_content.find('section', class_='news-content')

        user_question = content.find('strong')

        for i, tag in enumerate(content.find_all('h2')):
            question = tag
            answer = ''
            for tag in question.next_siblings:
                if tag.name == 'p' or tag.name == 'blockquote':
                    answer += tag.text
                if tag.name == 'h2':
                    break
            if i == 0:
                data.append({'title': title, 'user_question': user_question.text, 'question': question.text, 'answer': answer})
            else:
                data.append({'title': title, 'user_question': question.text, 'question': question.text, 'answer': answer})

        

        yield data
```

```python
import json

data_full = []
for data in Get_article():
    data_full += data
```
